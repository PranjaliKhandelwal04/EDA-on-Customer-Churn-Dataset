```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Customer Churn.csv')
df
```

```python
df.head(5)
```

```python
df.info()
```

```python
#Replacing blanks with 0 as tenure is 0 and total charges are there.

df["TotalCharges"] = df["TotalCharges"].replace(" ","0")
df["TotalCharges"] = df["TotalCharges"].astype("float")
```

```python
df.info()
```

```python
df.isnull().sum().sum()
```

```python
df.describe()
```

```python
df["customerID"].duplicated().sum()
```

```python
def conv(value):
  if value == 1:
    return "yes"
  else:
    return "no"

df["SeniorCitizen"] = df["SeniorCitizen"].apply(conv)
```

```python
df.head(30)
```

```python
ax = sns.countplot(x = "Churn",data = df)
plt.title("Count of customers by Churn",fontsize = 15)
ax.bar_label(ax.containers[0])
plt.show()
```

```python
gb = df.groupby("Churn").agg({"Churn":"count"})
gb
```

```python
plt.figure(figsize = (3,4))
plt.title("Percentage of customers by Churn",fontsize = 10)
plt.pie(gb["Churn"],labels = gb.index,autopct = "%1.2f%%")
plt.show()
```

```python
#From the given pie chart it has been cleared that 26.54% of the customers have churned out
#Not lets know the reason behind it
```

```python
plt.figure(figsize = (4,4))
sns.countplot(x = "gender",data = df, hue = "Churn")
plt.title("Churn by Gender",fontsize = 10)
plt.show()
```

```python
plt.figure(figsize = (4,4))
ab = sns.countplot(x = "SeniorCitizen",data = df)
ab.bar_label(ab.containers[0])
plt.title("Count of Customers by Senior Citizen",fontsize = 10)
plt.show()
```

```python
# Create a crosstab of counts
ct = pd.crosstab(df["SeniorCitizen"], df["Churn"], normalize="index") * 100  # percentage by row

# Plot stacked bar
ax = ct.plot(kind="bar", stacked=True, figsize=(5,5), color = ["green","yellow"])

# Add % labels
for c in ax.containers:
    ax.bar_label(c, fmt="%.1f%%", label_type="center", fontsize=9, color="black")

plt.title("Churn by Senior Citizen (Stacked Bar Chart)", fontsize=10)
plt.ylabel("Percentage(%)")
plt.xlabel("Senior Citizen")
plt.legend(title="Churn",bbox_to_anchor = (0.9,0.9))
plt.show()
```

```python
#Comparatively a greater percentage of senior citizen category have churned
```

```python
plt.figure(figsize = (8,4))
sns.histplot(x = "tenure",data = df,bins = 72,hue = "Churn")
plt.show()
```

```python
cols = [
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies'
]

# Define subplot grid size (3 rows × 3 columns for 9 plots)
n_cols = 3
n_rows = (len(cols) + n_cols - 1) // n_cols  # auto compute rows

fig, axes = plt.subplots (n_rows, n_cols, figsize=(20,n_rows*4))
axes = axes.flatten()  # flatten so we can index easily

# Loop through columns and plot
for i, col in enumerate(cols):
    sns.countplot(data=df, x=col, hue="Churn", ax=axes[i], palette="Set2")
    axes[i].set_title(f"{col} vs Churn", fontsize=10)
    axes[i].set_xlabel("")
    axes[i].set_ylabel("Count")

# Hide any unused subplots (if grid > number of columns)
for j in range(i+1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize = (6,4))
ac = sns.countplot(x = "PaymentMethod",data = df, hue = "Churn")
ac.bar_label(ac.containers[0])
ac.bar_label(ac.containers[1])
plt.title("Churn by Payment Method",fontsize = 10)
plt.legend(title="Churn",bbox_to_anchor = (0.9,0.9))
plt.xticks(rotation = 45)
plt.show()
```
